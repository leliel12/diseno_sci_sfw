### Lenguajes dinmicos y estticos

- Decimos que un lenguaje es de **tipado esttico**, porque los tipos tienen que definirse en tiempo de compilacin para que el programa funcione.
- Los lenguajes de **tipado dinmico**, definen los tipos de las variables se definen en tiempo de ejecucin.

> Como bien me indica [Juan Quijano por Twitter](https://twitter.com/jc_quijano/status/862264185278869506), esto no es realmente as, ya que hay lenguajes dinmicos compilados y estticos interpretados. Una definicin ms exacta sera la de que el valor de una variable est restringido al tipo que esa variable especifica en su definicin. En el caso de los lenguajes dinmicos, el tipo est asociado con valores en tiempo de ejecucin y no con nombres de variables.

### Lenguajes dinmicos y estticos

#### Tipados fuertes o tipados dbiles

- Otra cosa que suele confundirse aqu, es el tipado fuerte, o el tipado dbil. Tanto `Python` como `Bash`, son lenguajes dinmicos.
- `Python` tiene un tipado mucho ms fuerte que `Bash`. 
- Si sumas `1 + "1"` en Python, recibes un error en tiempo de ejecucin.
- Pero `bash` permite estas cosas:

```bash
$ a=2;
$ echo $(($a + 1)); 
3
$ echo $a hola
2 hola
```

### Lenguajes dinmicos y estticos
----

#### Lenguajes de tipado esttico

##### Supuestas Ventajas:

- Detencin temprana de algunos tipos de errores.
- Es ms sencillo refactorizar nombres de variables, funciones, mtodos etc.
- En general el cdigo es ms expresivo, y muchas veces es fcil entender lo que se est haciendo gracias a los tipos.
- Mejor autocompletado del cdigo. Como el compilador tiene ms informacin, es ms fcil que te ayude mientras ests escribiendo, ganando en productividad.

### Lenguajes dinmicos y estticos
----

#### Lenguajes de tipado esttico

##### Supuestas Desventajas:

- Compilar es lento. Y si tienes mucho cdigo, es muy lento.
- Tienes que invertir tiempo en aprender el sistema de tipos.
- A veces los errores que se producen con los tipos, son difciles de solucionar, porque el compilador no es demasiado claro.
- Estar atado a los tipos, hace que tengas ms restricciones.

### Lenguajes dinmicos y estticos
----

#### Lenguajes de tipado dinmico

##### Supuestas Ventajas:

- No pierdes el tiempo compilando. De hecho hacer TDD tiene un flujo mucho ms gil, ya que no tienes que compilar.
- Aunque no hay compilador, muchos lenguajes dinmicos tienen linterns que te ayudan a cazar muchos errores.
- Son lenguajes un poco ms interactivos. Suelen tener REPL que te permiten probar el cdigo fcilmente. Algunos lenguajes estticos tambin tienen, como C#, pero no son tan tiles.
- El cdigo es ms flexible.

### Lenguajes dinmicos y estticos
----

#### Lenguajes de tipado dinmico

##### Supuestas Desventajas:

- Hay errores que un compilador te detectara, pero que con un lenguaje dinmico te vas a tragar.
- Quieres cambiar el nombre a una funcin? Preparate para buscar bien en todo el cdigo (y ms si no tienes tests).
- Que los tipos sean dinmicos, no quiere decir que no te tengas que preocupar por ellos. De hecho no tienes tipos que te guen mientras codificas, as que tienes que tener en mente de qu tipo es cada variable, o que tipo devuelve una funcin. Y no siempre es sencillo.
