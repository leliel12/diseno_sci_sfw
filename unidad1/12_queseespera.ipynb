## Bag of coordinates.

- En general se estructuran alrededor de alguna idea que extiende de manera mucho mas compleja el concepto de vector.
- numpy esencialmente es la versión mas primitiva de este tipo de proyectos.
- Suelen presentar algún objeto principal que presenta operaciones complejas alrededor de datos de múltiples dimensiones.

**Salidos de acá:** Galaxychop, SpaceWeather.

## Model-fit

- Presentan objetos que se configuran con parámetros libres y luego ajustan a algún tipo estado interno, o crean algún nuevo objeto con el resultado del ajuste.
- Scikit-learn es el arquetipo.
- En general ofrecen una familia de modelos.

**Salidos de acá:** Galaxychop, Pyedra, caos, neuro. Arby.

## DataFrame-centric.

- Son una particularidad del bag-of-coordinates pero esta mucho mas ceñido alrededor de la idea de los datos son tablas, donde las observaciones se presentan en forma de filas y de cada observación se posee una serie de características fijas representadas por las columnas.
- Naturalmente el arquetipo es todo el ecosistema creado alrededor de Pandas (geopandas).
- Suele ser útil sacar estadísticas por columnas o grupos de filas.

**Salidos de acá:** Pyedra, APoDe, Garpar, Stratopy

## El wrapper.

- Envuelven un proyecto existente con el único objetivo de traer las funcionalidades a un nivel semántico mas cerca de la ciencia.
- Este nivel semántico puede ser desde exponer en el lenguaje actual cosas de otros lenguajes o de plataformas remotas para aprovecharlas con el ecosistema de python.
- A nivel diseño suelen ser super simples y mas preocupados en disminuir o tiempos de cómputo o de desarrollo.

**Salidos de acá:** Deepspyce, pycf3 (este es mío personal)

## Function-object

- Generan un objeto complejo a partir de una función.
- Cualquier simulación esencialmente sigue este arquetipo.
- Casi todos los proyectos tienen pequeños generadores de datos de juguete automáticos, pero hay grandes desarrollos construidos alrededor de este concepto.

**Salidos de aca:** Pyedra, Garpar

## Scientific - Kit

- No presentan solamente un modelo, sino una infraestructura para experimentar la construcción de modelos y experimentos de algún tipo.
- Proveen objetos, clases, funciones y constantes útiles para la creación de experimentos.
- Scikit-learn, Scikit-image y demás scikits son los mejores ejemplos.
- Suelen estar super especializados en un problema muy particular.

**Salidos de acá:** Pyrrha, y sospecho que neuro, garpar, python4fluis y caos tiene un potencial para crear un scikit.

## ?

- Hay librerías que simplemente presentan un montón de utilidades aplicables a un contexto, pero sin interfaces uniformes.
- Cada módulo es particularmente distinto de los demás.
- Si los scikits son navajas suizas estas son cajas de herramientas donde cada componente es muy diferente a los demás.
- Suelen representar problemas muy fundamentales de algún dominio.
- Se me ocurren: Scipy y Astropy.
