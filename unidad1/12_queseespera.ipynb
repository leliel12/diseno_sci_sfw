## Bag of coordinates.

- En general se estructuran alrededor de alguna idea que extiende de manera mucho mas compleja el concepto de vector.
- numpy esencialmente es la versin mas primitiva de este tipo de proyectos.
- Suelen presentar algn objeto principal que presenta operaciones complejas alrededor de datos de mltiples dimensiones.

**Salidos de ac:** Galaxychop, SpaceWeather.

## Model-fit

- Presentan objetos que se configuran con parmetros libres y luego ajustan a algn tipo estado interno, o crean algn nuevo objeto con el resultado del ajuste.
- Scikit-learn es el arquetipo.
- En general ofrecen una familia de modelos.

**Salidos de ac:** Galaxychop, Pyedra, caos, neuro. Arby.

## DataFrame-centric.

- Son una particularidad del bag-of-coordinates pero esta mucho mas ceido alrededor de la idea de los datos son tablas, donde las observaciones se presentan en forma de filas y de cada observacin se posee una serie de caractersticas fijas representadas por las columnas.
- Naturalmente el arquetipo es todo el ecosistema creado alrededor de Pandas (geopandas).
- Suele ser til sacar estadsticas por columnas o grupos de filas.

**Salidos de ac:** Pyedra, APoDe, Garpar, Stratopy

## El wrapper.

- Envuelven un proyecto existente con el nico objetivo de traer las funcionalidades a un nivel semntico mas cerca de la ciencia.
- Este nivel semntico puede ser desde exponer en el lenguaje actual cosas de otros lenguajes o de plataformas remotas para aprovecharlas con el ecosistema de python.
- A nivel diseo suelen ser super simples y mas preocupados en disminuir o tiempos de cmputo o de desarrollo.

**Salidos de ac:** Deepspyce, pycf3 (este es mo personal)

## Function-object

- Generan un objeto complejo a partir de una funcin.
- Cualquier simulacin esencialmente sigue este arquetipo.
- Casi todos los proyectos tienen pequeos generadores de datos de juguete automticos, pero hay grandes desarrollos construidos alrededor de este concepto.

**Salidos de aca:** Pyedra, Garpar

## Scientific - Kit

- No presentan solamente un modelo, sino una infraestructura para experimentar la construccin de modelos y experimentos de algn tipo.
- Proveen objetos, clases, funciones y constantes tiles para la creacin de experimentos.
- Scikit-learn, Scikit-image y dems scikits son los mejores ejemplos.
- Suelen estar super especializados en un problema muy particular.

**Salidos de ac:** Pyrrha, y sospecho que neuro, garpar, python4fluis y caos tiene un potencial para crear un scikit.
