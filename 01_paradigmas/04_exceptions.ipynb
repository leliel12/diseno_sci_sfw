## Exceptions

- El manejo de excepciones es el proceso de responder a la ocurrencia de excepciones: condiciones anmalas o excepcionales que requieren un procesamiento especial durante la ejecucin de un programa.
- Una excepcin interrumpe el flujo normal de ejecucin y ejecuta un manejador de excepciones previamente registrado.
- El manejo de excepciones, si se proporciona, se facilita mediante construcciones especializadas del lenguaje de programacin, mecanismos de hardware como las *interrupciones* o las facilidades de comunicacin entre procesos del sistema operativo (SO), como *signals*.
- Algunas excepciones, especialmente las de hardware, pueden reanudarse donde se interrumpi.


## Exceptions

- La definicin de una excepcin se basa en la observacin de que cada procedimiento tiene una precondicin, o un conjunto de circunstancias para las cuales terminar "normalmente".
- Un mecanismo de manejo de excepciones permite que el procedimiento lance una excepcin si esta precondicin se viola:
    - Por ejemplo, si se llama al procedimiento con un conjunto anormal de argumentos.
- El mecanismo de manejo de excepciones luego se encarga de la excepcin.
- La precondicin y la definicin de excepcin son subjetivas. El conjunto de circunstancias "normales" est definido por el programador.
- Las excepciones son una solucin al ["problema del semipredicado"](https://en.wikipedia.org/wiki/Semipredicate_problem)

Exceptions - `AssertionError`
---

- En lugar de esperar a que un programa se rompa en producttion, se puede usar la palabra reservada `assert`. 
- Afirmamos que se cumple una determinada condicin.
- Si la condicin resulta ser falsa, el programa arroja una excepcin `AssertionError`.
- Los `assert` se omiten cuando se ejecuta Python en modo optimizado `python -O`


```python
import sys
assert ('linux' not in sys.platform), "This code can't run on Linux systems."
```

## Exceptions - Herencia
----
Podemos crear excepciones personalizadas ms especficas.

```python
class InvalidPixelValueError(ValueError):
    pass
```

```python
data = np.array([[1., 1., 1.],
                 [1., 1., np.nan],
                 [1., 1., 1.]])
```

```python
if not np.isfinite(data).all():
    raise InvalidPixelValueError('Hay un valor invalido')
```

### Nuevo comportamiento
Podemos agregar comportamiento ms especfico

```python
class InvalidPixelValueError(ValueError):
    
    def __init__(self, idx, message='Hay un valor invalido'):
        self.idx = idx
        self.message = message
        super().__init__(self.message)

    #def __str__(self):
    #    return f'{self.message} --> {self.idx}'
```

```python
def validar(array):
    bad_idx = np.argwhere(~np.isfinite(array))
    return bad_idx
```

```python
if (idx := validar(data)).size:
    raise InvalidPixelValueError(idx, 'Hay un valor invalido')
```
