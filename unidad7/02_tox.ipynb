# Archivo de configuración: tox.ini
----
Leer la documentación: https://tox.readthedocs.io/en/latest/

```ini
# content of: tox.ini , put in same dir as setup.py
[tox]
envlist = py39, py37, style, coverage, check-manifest

[testenv]
deps = pytest
commands = pytest tests/ {posargs}

[testenv:style]
deps =
    flake8
    flake8-black
    flake8-isort
commands = flake8 proyectox/ tests/ {posargs}

[testenv:coverage]
deps =
    coverage
    pytest-cov
commands =
    pytest tests/ --cov=proyectox/ --cov-append --cov-report=term-missing
    --fail-under=80

[testenv:check-manifest]
deps = check-manifest
commands = check-manifest
```

# Cuándo hacer commit?
----
A partir de ahora, **Tox es su principal herramienta de trabajo**. Les permite desarrollar código en forma fluida confiando en su suite de tests. Si introducimos modificaciones y tox muestra errores (test, estilo, coverage), volvemos atrás y corregimos. Cuando está todo en verde podemos continuar desarrollando.
