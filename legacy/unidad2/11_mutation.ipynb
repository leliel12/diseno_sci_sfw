## Qu es el mutation testing?

- Tenemos nuestro proyecto + los tests.
- Introducimos **un** solo cambio al proyecto.
- Corremos nuevamente los tests.
- Idealmente cualquier cambio llevara a la falla en los tests.

```python
for operator in mutation_operators:
    for site in operator.find_sites(code):
        operator.mutate(site)
        run_tests
```

## Que nos responde un test de mutacin?

- **Killed:** El test detecto la mutacin.
- **Incompetent:** La mutacin produjo codigo que es inherentemente incorrecto.
- **Survived:**
    - Los tests fallaron en detectar el cambio.
    - **O** los test son inadecuado para esa fraccin de cdigo.
    - Codigo mutado es equivalente.
    


# Objetivo de Mutation Testing

1. **Coverage analysis:** Mis tests cubren significativamente mi cdigo
2. Diferenciar el concepto de linea ejecutada vs funcionalidad verificada.

Es buena idea partir del 100% de code-coverage


## Mutations examples:
    
- Remplaazar un operador relacional: `x > 1` -> `x >= 1`
- Reemplazar `break` por `continue`
- Reemplazar una condicin logica (`and` -> `or`)
- Cambiar una constante.
- Modificar un string.
- Borrar un decorador.

## Complejidad 1: La velocidad

- Esto es lento!... MUY LENTO
- Hay que tratar de que todo vaya en parelelo.
- Luego de una primer corrida de baseline: 
    - Solo correr los mutantes que fallaron.
    - Solo mutar codigo modificado.
- Acelerar los tests.
- Frenar los test al primer fallo.

## Complejidad 3: Mutantes equivalentes

````python
esta_50 = False
for e in range(500):
    if e == 50:
        esta_50 = True
        break
        
assert esta_50
````

````python
esta_50 = False
for e in range(500):
    if e == 50:
        esta_50 = True
        continue  # Mutacin
        
assert esta_50
````



Mutmut is a mutation testing system for Python, with a strong focus on ease of use.

Some highlight features:

- Found mutants can be applied on disk with a simple command making it very easy to work with the results
- Remembers work that has been done, so you can work incrementally
- Can use coverage data to only do mutation testing on covered lines
- Battle tested on real libraries by multiple companies
