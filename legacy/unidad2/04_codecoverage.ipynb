## Unidad 2: Covertura de cdigo




## Problemas con el testeo

- Es muy fcil escribir tests que no hagan nada (de hecho podemos escribir 100 funciones vacias y mostrar que los tests pasan)
- La cobertura de cdigo es una medida (porcentual) en las pruebas de software que mide el grado en que el cdigo fuente de un programa ha sido comprobado.
- Es una forma indirecta de calcular la calidad de un test.
- Aun asi puede engaarse al coverage.

## Coverage.py

Coverage.py es una herramienta para medir la cobertura de cdigo de los programas Python. Monitorea su programa, observando qu partes del cdigo se han ejecutado, luego analiza la fuente para identificar el cdigo que podra haberse ejecutado pero que no.

La medicin de cobertura generalmente se usa para medir la efectividad de las pruebas. Puede mostrar qu partes de su cdigo estn siendo ejercitadas por las pruebas y cules no.

https://coverage.readthedocs.io

## Vamos con un ejemplo bsico

```python
# contenido  test_cov.py
def division(a, b):
    return a / b
def sum(a, b):
    return a + b
def test_division():
    assert division(1, 2) == .5
def test_sum():
    assert sum(1, 2) == 3
if __name__ == "__main__":
    test_division()
    test_sum()
```

```python
!coverage run 04_code_coverage_code/test_cov.py
!coverage report -m
```

## Integrandolo con pytest

```python
!pytest -v 04_code_coverage_code/test_cov.py --cov test_cov --cov-fail-under 90 --cov-report term-missing
```

```python
!coverage report -m
```

## Test-Driven Development
-----

TDD es una disciplina de desarrollo que propone escribir los test unitarios ANTES de escribir el cdigo real.

<table >
	<tbody>
		<tr>
			<td width=50%></td>
            <td><b>Segun <i>Uncle Bob</i>, TDD se puede describir en tres leyes:</b><br>
                
1. You are not allowed to write any production code unless it is to make a failing unit test pass.<br>     
2. You are not allowed to write any more of a unit test than is sufficient to fail; and compilation failures are failures.<br>
3. You are not allowed to write any more production code than is sufficient to pass the one failing unit test.</td></tr>
